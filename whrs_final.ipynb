```python
import calendar
import pickle
```

```python

# functions for frame C

#function for date prediciton
def pred_date():
    with open('model_cycle_len','rb') as g:
        model=pickle.load(g)
    data=pd.read_csv('user_data.csv')
    ans=model.predict(data.drop('LengthofCycle',axis=1))
    #cycle_len=user_data['MeanCycleLength'][0]

    return str(ans[0])


def pred_phase():
    phase= "Menstrual phase"
    return phase

def pred_ovu():

    with open('model_ovulation','rb') as g:
        model=pickle.load(g)
    data=pd.read_csv('user_data.csv')
    ans=model.predict(data.drop('EstimatedDayofOvulation',axis=1))
    
    return str(ans[0])


    
```

```python

# FRAME C
#_____________________________________________________________________________________________ 
# frame c: prediction frame
pred_frame= tkinter.LabelFrame (frame,  width= 100, height= 100,text= "Prediction" , bg= "#ffb6c1" , font= 20)
pred_frame.grid(row=0,column=1,padx=40,pady=40)


next_date= pred_date()+ " days"
pred_phase= pred_phase()
pred_ovu=pred_ovu()+" days"

#objects of frame C

# predicted date

next_date_label = tkinter.Label(pred_frame, text="Predicted length of ongoing cycle: ", bg='#ffb6c1')
next_date_label.grid(row=0, column=0)

next_date_out = tkinter.Label(pred_frame, text=next_date, bg='#FFFFFF')
next_date_out.grid(row=1, column=0,padx=10,pady=10)

# predicted phase

pred_phase_label = tkinter.Label(pred_frame, text="Predicted Phase (current): ", bg='#ffb6c1')
pred_phase_label.grid(row=2, column=0)

pred_phase_out = tkinter.Label(pred_frame, text=pred_phase, bg='#FFFFFF')
pred_phase_out.grid(row=3, column=0,padx=10,pady=10)

# ovulation phase

pred_phase_label = tkinter.Label(pred_frame, text="Predicted time for ovulation: ", bg='#ffb6c1')
pred_phase_label.grid(row=4, column=0)

pred_phase_out = tkinter.Label(pred_frame, text=pred_ovu, bg='#FFFFFF')
pred_phase_out.grid(row=5, column=0,padx=10,pady=10)

```
